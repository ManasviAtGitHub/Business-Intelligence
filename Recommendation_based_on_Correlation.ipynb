```python
rating = pd.DataFrame(frame.groupby('placeID')['rating'].mean())
rating.head()
```

```python
rating['rating_count'] = pd.DataFrame(frame.groupby('placeID')['rating'].count())
rating.head()
```

```python
places_crosstab = pd.pivot_table(data=frame, values='rating', index='userID', columns='placeID')
places_crosstab.head()
```

```python
places_crosstab.columns
```

```python
places_crosstab.loc[:, [135076, 135085]].head(20)
```

```python
TLH_ratings = places_crosstab[135085]
TLH_ratings[TLH_ratings>=0]
```

```python
%%capture
similar_to_TLH = places_crosstab.corrwith(TLH_ratings)

corr_TLH = pd.DataFrame(similar_to_TLH, columns=['PearsonR'])
corr_TLH.dropna(inplace=True)
```

```python
corr_TLH.head()
```

```python
TLH_corr_summary = corr_TLH.join(rating['rating_count'])
```

```python
TLH_corr_summary.head()
```

```python
TLH_corr_summary.sort_values('PearsonR', ascending=False).head(10)
```

```python
TLH_corr_summary[TLH_corr_summary['rating_count']>=10].sort_values('PearsonR', ascending=False).head(10)
```
